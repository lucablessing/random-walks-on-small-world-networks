# RANDOM WALK ON A 2D LATTICE WITH PERIODIC BOUNDARY CONDITIONS

```python
import networkx as nx
import matplotlib.pyplot as plt
from random import randrange

L=9   #linear dimension of the 2d graph
T=100 #number of random step
G = nx.grid_2d_graph(L,L, periodic=True)
G.remove_edges_from(list(G.edges)) # remove edges

#starting point the origin (0,0)
v=0
w=0
for t in range(0,T-1):
    #this variables have the memory of the previous vertex
    x=v
    y=w
    walk = randrange(0, 3) #random integer number between 0,3
    if walk==0:
        if v==L-1 : #boundary condition
            v=0
        else :
            v=v+1
        
    elif walk==1:
        if v==0 : #boundary condition
            v=L-1
        else :
            v=v-1
    elif walk==2:
        if w==L-1 : #boundary condition
             w=0
        else :
             w=w+1
    else:
        if w==0 : #boundary condition
             w=L-1
        else :
             w=w-1
    G.add_edge((x,y),(v,w)) # at each step I add an undirectional edge
    
#PLOT
color_map = []
for node in G:
    if node == (0,0):
        color_map.append('red')
    elif node == (v,w): 
        color_map.append('blue') 
    else:
        color_map.append('lightgreen')
plt.figure(figsize=(6,6))
pos = {(x,y):(y,-x) for x,y in G.nodes()}
nx.draw(G, pos=pos, 
    node_color=color_map, 
    with_labels=False,
    node_size=100)
```
